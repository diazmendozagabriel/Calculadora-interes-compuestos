```python
!git status
# 1. LA FUNCIÓN (La calculadora pura)
def interes_compuesto(CI, I, AÑOS):
    formula = CI * (1 + I) ** AÑOS
    return formula  # Devuelve el resultado de la operación matemática


# 2. LAS PREGUNTAS

CI = float(input('Introduce tu capital inicial ($): '))
I = float(input('Cual va a ser la tasa de interes (ej: 0.05 para 5%): '))
AÑOS = int(input('Cuantos años vas a invertir: '))

# 3. LA LLAMADA Y EL RESULTADO
# Ahora que tenemos las respuestas guardadas en las variables, se las enviamos a la función
resultado = interes_compuesto(CI, I, AÑOS)

print(f"El capital final sera {resultado:.2f} $")
```
