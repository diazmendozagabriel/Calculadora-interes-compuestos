```python
#Funciona como el IF de Excel
concierto= True
if concierto:
    print ("acho PR es otra cosa")
else:
    print ("me quedo en casa estudiando")
#Nota= elif: Es el punto medio, una abreviatura de "si no, si...".
#Por ejemplo: Si hay leche, café con leche. Si no, si (elif) hay limonada, me tomo un vaso. Si no (else), bebo agua.
```

```python
ingredientes = ["harina", "huevos", "leche"]

# 'ingrediente' (en singular) es una variable temporal que toma 
# el valor de cada elemento en cada vuelta del bucle
for ingrediente in ingredientes:
    print("Añadiendo a la mezcla: " + ingrediente)
```

```python
temperatura_sopa = 10  # Empieza fría

while temperatura_sopa < 40:
    print("La sopa sigue fría... revolviendo.")
    temperatura_sopa = temperatura_sopa + 10  # Calentamos la sopa en cada vuelta

print("¡La sopa ya está caliente! A comer.")
```

```python
# 'nombre' e 'ingrediente' son los argumentos que la función espera recibir
def preparar_plato(nombre_cliente, ingrediente):
    print("Marchando un plato para " + nombre_cliente)
    print("Cocinando con: " + ingrediente)
    print("---")

# Al llamar a la función, le pasamos los datos reales
preparar_plato("Carlos", "papas")
preparar_plato("Ana", "pasta")
```

```python
# Una función matemática que calcula el precio con IVA
def calcular_total(precio_base):
    precio_final = precio_base * 1.21  # Sumamos el 21% de IVA
    return precio_final  # Devolvemos el resultado al exterior

# Guardamos lo que la función "nos devuelve" en una variable
total_compra = calcular_total(100)
print(total_compra)  # Imprimirá: 121.0
```
